# Backtest Analysis for the file generated by momentum scanner (WIP)

```python
import pandas as pd
from xlrd import open_workbook
import pprint
pp = pprint.PrettyPrinter(indent=4)

ENTRY_RANK = 2
EXIT_RANK = 20
```

### Enter the path name of the file to be analysed

```python
input_file = r'data/output/momentum_monthly_output_sample.xlsx'
```

#### Read list of sheets in excel file and remove unnecessary sheet.

```python
xl = open_workbook(input_file)
sheets = xl.sheet_names
sheets.remove("Sheet1")
sheets.sort()
num_of_sheets = len(sheets)
print(f"Total number of sheets in excel file: {num_of_sheets}")
print(sheets)
```

```python
df = pd.DataFrame()
```

```python
final_dict = {}
wb = open_workbook(input_file) 
```

```python
for sheet in sheets:
    sh = wb.sheet_by_name(sheet)  
    for i in range(1,ENTRY_RANK+1):
        d = {}
        stock = sh.cell(i,0).value
        d['entry_date'] = sheet
        d['stock_ltp'] = sh.cell(i,2).value
        d['exit_date'] = sheet
        d['exit_ltp'] = None
        final_dict[stock] = d
```

```python
pp.pprint(final_dict)
```

```python
df = pd.DataFrame(final_dict)
df = df.transpose()
```
